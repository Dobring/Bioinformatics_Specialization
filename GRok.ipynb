```python
def freq_tab(dna, t):
    tmer = {}
    freq_pat = []
    for i in range(len(dna)-t):
        if dna[i:i+t] not in tmer:
            tmer[dna[i:i+t]] = 1
        else:
            tmer[dna[i:i+t]] = tmer[dna[i:i+t]] + 1
    return tmer
```

```python
def CountMatch(pat, dna, H):
    count_match = 0
    lenpat = len(pat)
    for i in range(len(dna) - lenpat+1):
        if HamDis(dna[i:i+lenpat], pat) <= H:
            count_match += 1
    return count_match
```

# Generating the Neighborhood of a String


```python
def ImmedNei(pat):
    nei = [pat]
    for i in range(len(pat)):
        symbol = pat[i]
        for x in 'ACGT':
            if x != symbol:
                neighbor = pat[:i] + x + pat[i+1:]
                nei.append(neighbor)
    return nei
```

```python
ImmedNei('AAAA')
```

```python
def Neighbors(pattern, d):
    if len(pattern) == 0:
        return pattern
    if len(pattern) == 1:
        return ['A', 'C', 'G', 'T']
    neighborhood = []
    SuffixNeighbors = Neighbors(pattern[1:], d)
    for i in SuffixNeighbors:
        if HamDis(pattern[1:], i) < d:
            for x in 'ACGT':
                neighbor = x + i
                neighborhood.append(neighbor)
        else:
            neighbor = pattern[0] + i
            neighborhood.append(neighbor)
    return neighborhood
```

```python
def FreqWWMism(dna, k, d):
    patterns = []
    freqmap = {}
    n = len(dna)
    for i in range(n-k):
        pattern = dna[i:i+k]
        neighborhood = Neighbors(pattern, d)
        for j in neighborhood:
            if j not in freqmap:
                freqmap[j] = 1
            else:
                freqmap[j] += 1
    m = max(freqmap.values())
    for i in freqmap.keys():
        if freqmap[i] == m:
            patterns.append(i)
    return freqmap
```
